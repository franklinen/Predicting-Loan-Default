### Preprocessing

##### Remove missing values

```python
#use lambda function to fill missing values with the most frequent vales for the column
df = df.apply(lambda x:x.fillna(x.value_counts().index[0]))
```

```python
#check for missing value again
df.isnull().sum()
```

##### Rename Feature Columns

Our feature columns have different "capitalizations" in their names, e.g. both upper and lower "case". We remove all spacesfor feature columns and rename to shorter column names.



```python
df.rename(columns = { 'Credit_Limit': 'creditlim', 'Sex':'sex', 'Education':'education','Marital_Status': 'maritalstat', 'Age': 'age', 'Jan_Repay_Status': 'janstatus', 'Previous_Payment_Prior_Jan': 'janpay', 'Jan_Statement': 'janstatement', 'Feb_Repay_Status':'febstatus', 'Previous_Payment_Prior_Feb': 'febpay', 'Feb_Statement': 'febstatement', 'Mar_Repay_Status':'marstatus', 'Previous_Payment_Prior_Mar': 'marpay', 'Mar_Statement': 'marstatement', 'Apr_Repay_Status':'aprstatus', 'Previous_Payment_Prior_Apr': 'aprpay', 'Apr_Statement': 'aprstatement', 'May_Repay_Status':'maystatus', 'Previous_Payment_Prior_May': 'maypay', 'May_Statement': 'maystatement', 'Jun_Repay_Status':'junstatus', 'Previous_Payment_Prior_Jun': 'junpay', 'Jun_Statement': 'junstatement', 'July_Payment_Status':'julstatus'}, inplace = True) 

# Output the first two rows.
df.head(2)
```

##### convert categorical colums to integer from float


```python
cols = ['sex','education','maritalstat','age','janstatus','febstatus','marstatus','aprstatus','maystatus','junstatus','julstatus']
for col in cols:
    df[col] = df[col].apply(lambda x: int(x) if x == x else "")


df.head(n=5)
```

## Modeling

##### import Classifiers and assign to variable

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC

model1 = LogisticRegression()
model2 = RandomForestClassifier()
model3 = AdaBoostClassifier(n_estimators=5)
model4 = KNeighborsClassifier()
model5 = SVC(kernel='linear')
```

##### Modeling Preprocessing

split data into X and Y (to extract our target variable from the rest of the variables)

```python
#split data into X and Y
X=df1.drop(['julstatus'],axis=1)
y=df1.julstatus
X.head()
```

import normalisation package for scaling(Scaling the features to reduce error from )

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_scaled = sc.fit_transform(X)
```

Run RFE and select best subset of X (Feature selection method.Used logistic regression for the selection)

```python
max_accu = 0
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import RFE
estimator = LogisticRegression() #use regression model for regression problem
list_r2=[]
max_r2=0
for i in range(1,len(X.iloc[0])+1):
    selector = RFE(estimator, i, step=1)
    selector = selector.fit(X_scaled, y)
    accuracy = selector.score(X_scaled,y)
    if max_accu < accuracy:
        sel_features = selector.support_
        max_accu = accuracy
        
X = X_scaled[:,sel_features]
print(X)
```

##### Modeling

Split the data into training set and testing set using train_test_split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

Train and fit different regression model on the training set.

```python
model1.fit(X_train,y_train)
model2.fit(X_train,y_train)
model3.fit(X_train,y_train)
model4.fit(X_train,y_train)
model5.fit(X_train,y_train)
```

##### Predictions and Evaluations

predictions for test

```python
y_pred1 = model1.predict(X_test)
y_pred2 = model2.predict(X_test)
y_pred3 = model3.predict(X_test)
y_pred4 = model4.predict(X_test)
y_pred5 = model5.predict(X_test)
```

check accuracy. Classifier performance on test

```python
print(model1.score(X_test,y_test)) #Logistic Regression
print(model2.score(X_test,y_test)) #Random Forest Classifier
print(model3.score(X_test,y_test)) #AdaBoost Classifier
print(model4.score(X_test,y_test)) #KNearest Neighbors Classifier
print(model5.score(X_test,y_test)) #Support Vector Classifier
```

Create a classification report for the models

GridSearch to select best model parameters
